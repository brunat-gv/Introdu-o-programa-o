```python

"""1 — Escreva um programa que:
Peça um número inteiro ao usuário. Imprima todos os números de 1 até esse número usando um while."""

num = int(input("Digite um número inteiro: "))
i = 1
while i <= num:
    print(i)
    i += 1

#_______________________________________________________________________________________

"""2 — Escreva um programa que:
Peça um número inteiro. Imprima todos os números pares de 0 até esse número usando um for."""

num = int(input("Digite um número inteiro: "))
for i in range(0, num + 1, 2):
   print(i)

#_______________________________________________________________________________________

"""3 — Escreva um programa que:
Peça 5 números ao usuário. Ao final, mostre a soma desses números.
Obrigatório: usar um laço de repetição."""

soma = 0
for i in range (5):
    valor = int(input("Digite um números:" ))
    soma = soma + valor
print("Este é o resultado da soma desses valores",soma )

#_______________________________________________________________________________________

"""Questão 4 — Escreva um programa que:
Peça um número inteiro positivo. Mostre a tabuada desse número de 1 até 10."""

num = int(input("Digite o número desejado: "))

for i in range ( 1, 11):
    mult = num * i
    print(num,"X", i, "=",mult)

#_______________________________________________________________________________________

"""#Faça um programa que imprima os números de 1 até 10 usando while."""

cont = 1
while cont <= 10:
    print(cont)
    cont  = cont + 1

#________________________________________________________________________________________

"""Peça um número inteiro ao usuário.
Depois, imprima todos os números pares de 0 até esse número usando while."""

num = int(input("Digite um número inteiro: "))
cont = 0
while cont <= num:
    print(cont)
    cont = cont + 2
```
